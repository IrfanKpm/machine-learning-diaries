```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Step 1: Create Dummy Data
# Set seed for reproducibility
np.random.seed(42)

# Generate dummy data
n_samples = 1000
X = np.random.randn(n_samples, 2)  # Two features
y = (X[:, 0] + X[:, 1] > 0).astype(int)  # Binary target variable based on a simple rule
```

```python
# Create a DataFrame
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
df['Target'] = y

# Show the first few rows of the DataFrame
df.head()
```

```python
# Step 2: Visualize the Data
# Scatter plot of the dummy data
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature 1', y='Feature 2', hue='Target', data=df, palette='coolwarm')
plt.title('Scatter Plot of Dummy Data')
plt.show()
```

```python
# Step 3: Split the Data
# Define features and target variable
X = df[['Feature 1', 'Feature 2']]
y = df['Target']
```

```python
df.head(10)
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 4: Train the Model
# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Step 5: Evaluate the Model
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Print the classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
# Step 6: Plot Heatmaps
# Plot confusion matrix as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
```

```python
# Step 7: Predict New Data
# New data point
new_data = np.array([[1.601, 0.812]])

# Predict the class for the new data point
prediction = model.predict(new_data)
prediction_prob = model.predict_proba(new_data)

print(f"Prediction for new data point {new_data}: {prediction[0]}")
```

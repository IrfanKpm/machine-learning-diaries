```python
import pandas as pd

# GitHub raw URL for the CSV file
url = "https://raw.githubusercontent.com/IrfanKpm/Machine-Learning-Notes1/main/datasets/pokemon.csv"
## original source > https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6


# Read the CSV file into a DataFrame
df = pd.read_csv(url)

# Display the first few rows of the DataFrame to verify
df.head()
```

```python
# Check if "Type 1" column is equal to "Fire"
result = df["Type 1"] == "Fire"
result
```

```python
# Boolean indexing with .loc:
# Select all rows from the DataFrame df where the value in the "Type 1" column is equal to "Fire".
fire_pokemon = df.loc[df["Type 1"] == "Fire"]

fire_pokemon
```

```python
df.loc[df["Attack"] >170]
```

```python
# Sorting DataFrame by column "Name" in descending order and displaying the top 5 rows.
sorted_df = df.sort_values("Name", ascending=False).head(5)
sorted_df
```

```python
df.sort_values(["Name","HP"]).head(10)
```

```python
# trying to calculate a new column "Energy" in your DataFrame df, based on the sum of several existing columns
             #  ("HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed", and "Generation")
df["Energy"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Atk"] + df["Speed"] + df["Generation"]
df.head(5)
```

```python
new_df = df.drop(columns=["Type 1","Type 2"]).head(5) # column droppinf
new_df
```

```python
df.loc[ ( df["Attack"] >170 ) & ( df["Defense"] >100 ) ]  # Filtering ( & :: AND )
```

```python
spr = df.loc[ ( df["Attack"] >170 ) | ( df["Defense"] >100 ) ]  # Filtering ( | :: OR )
spr
```

```python
# reser index number for this set
spr = spr.reset_index()
spr
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
```

```python
# Create a list of 20 student names
names = ['John', 'Alice', 'Bob', 'Eve', 'Charlie', 'David', 'Grace', 'Hank', 'Isaac', 'Jack',
         'Kate', 'Leo', 'Mona', 'Nancy', 'Oscar', 'Paul', 'Quincy', 'Rose', 'Steve', 'Tom']

# Create random marks for Physics, Chemistry, and Maths between 20 and 100
np.random.seed(42)  # For reproducibility
physics_marks = np.random.randint(20, 101, size=20)
chemistry_marks = np.random.randint(20, 101, size=20)
maths_marks = np.random.randint(20, 101, size=20)

# Generate random 10-digit mobile numbers, set 6 of them to NaN
mobile_numbers = [f'9{np.random.randint(100000000, 999999999)}' for _ in range(14)] + [np.nan]*6
np.random.shuffle(mobile_numbers)  # Shuffle so NaNs are randomly distributed

# Create the DataFrame
df = pd.DataFrame({
    'Name': names,
    'Mobile': mobile_numbers,
    'Physics': physics_marks,
    'Chemistry': chemistry_marks,
    'Maths': maths_marks
})
```

```python
df["Passed"] = (df.Maths + df.Chemistry + df.Physics ) > 120
df
```

```python
df[df.Passed == True]
```

```python
# Calculate the mean for each subject
mean_physics = df['Physics'].mean()
mean_chemistry = df['Chemistry'].mean()
mean_maths = df['Maths'].mean()

# Calculate the standard deviation for each subject
std_physics = df['Physics'].std()
std_chemistry = df['Chemistry'].std()
std_maths = df['Maths'].std()

# Count the number of students who passed
passed_count = df['Passed'].sum()

# Print the overview
print(f"Mean Marks:")
print(f"Physics: {mean_physics:.2f}")
print(f"Chemistry: {mean_chemistry:.2f}")
print(f"Maths: {mean_maths:.2f}")
print()
print(f"Standard Deviation of Marks:")
print(f"Physics: {std_physics:.2f}")
print(f"Chemistry: {std_chemistry:.2f}")
print(f"Maths: {std_maths:.2f}")
print()
print(f"Total Students Passed: {passed_count} out of {len(df)}")
```

```python
import matplotlib.pyplot as plt

# Data for average marks
subjects = ['Physics', 'Chemistry', 'Maths']
mean_marks = [mean_physics, mean_chemistry, mean_maths]

# Bar plot for mean marks
plt.figure(figsize=(10, 5))
plt.bar(subjects, mean_marks, color=['blue', 'green', 'orange'])
plt.title('Average Marks in Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.ylim(0, 100)  # Set y-axis limit to 100
plt.grid(axis='y')
plt.show()
```

```python
# Data for standard deviation
std_devs = [std_physics, std_chemistry, std_maths]

# Bar plot for standard deviation
plt.figure(figsize=(10, 5))
plt.bar(subjects, std_devs, color=['blue', 'green', 'orange'])
plt.title('Standard Deviation of Marks in Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Standard Deviation')
plt.ylim(0, max(std_devs) + 10)  # Set y-axis limit
plt.grid(axis='y')
plt.show()
```

```python
# Pie chart for passed vs. failed
passed_failed = [passed_count, len(df) - passed_count]
labels = ['Passed', 'Failed']
colors = ['lightgreen', 'salmon']

plt.figure(figsize=(8, 8))
plt.pie(passed_failed, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Students Passed vs. Failed')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```
